```python
import pandas as pd
import os
from windrose import WindroseAxes
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2000 + n for n in range(14, 20)]

df = None
state = True
for year in years:
    temp_data = pd.read_excel(r'wind_data\NMBU_WIND_10min_{0}.xlsx'.format(year))
    if state:
        df = temp_data
        state = False
    df = df.append(temp_data)
```

```python
df
```

```python
wd = df['VR']
ws = df['VH']
from math import pi
ax = WindroseAxes.from_ax()
ax.bar(wd, ws, normed=True, bins=np.arange(0, 8, 1), nsector = 16)
plt.xticks(np.arange(0, 2*pi, pi/4), ('E', 'NE', 'N', 'NW', 'W', 'SW', 'S', 'SE'), fontsize=16)
ax.set_legend(loc=(0.95,0), fontsize=16)
plt.savefig('year_windrose.png')
```
