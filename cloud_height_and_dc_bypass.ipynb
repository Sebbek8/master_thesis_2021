```python
# Define endpoint and parameters
endpoint = 'https://frost.met.no/observations/v0.jsonld?sources=SN18700&referencetime=2021-04-13%2F2021-04-14&elements=cloud_base_height1%2C%20cloud_base_height2%2C%20cloud_base_height3%2C%20cloud_top_height1%2C%20cloud_top_height2%2C%20cloud_top_height3%2C%20cloud_area_fraction%2C%20cloud_area_fraction1'
parameters = {
    'sources': 'SN18700',
    'elements': 'HS1 – cloud_base_height1, HS2 – cloud_base_height2, HT1 – cloud_top_height1, NN – cloud_area_fraction, NS1-cloud_area_fraction1',
    'referencetime': '2021-04-12/2021-04-14',
}
# Issue an HTTP GET request
r = requests.get(endpoint, parameters, auth=(client_id,''))
# Extract JSON data
json = r.json()
```

```python
# Check if the request worked, print out any errors
if r.status_code == 200:
    data = json['data']
    print('Data retrieved from frost.met.no!')
else:
    print('Error! Returned status code %s' % r.status_code)
    print('Message: %s' % json['error']['message'])
    print('Reason: %s' % json['error']['reason'])
```

```python
import datetime as dt

time_column = df['time']

time_column = time_column.apply(lambda x: dt.datetime.strptime(x, '%Y%m%d:%H%M').strftime('%Y-%m-%d %H:%M'))
    
time_column = pd.to_datetime(time_column, format=f'%Y-%m-%d %H:%M')

df['time'] = time_column
df
```

```python
import matplotlib.pyplot as plt

def kde_approx(start, end):
    #interval = df.loc[df['total_radiation']>0]
    interval=df
    length = len(interval)
    interval['total_radiation'].plot(kind='kde', color='lime')
    sliced = interval.loc[(interval['total_radiation'] >= start) & (interval['total_radiation'] < end)]
    frac = len(sliced)/length
    plt.xlim(-50, 1100)
    plt.grid()
    plt.xlabel('Irradiation [W/m2]')
    #plt.axvline(start, ls='--', alpha=0.4)
    #plt.axvline(end, ls='--', alpha=0.4)
    energy = 365*24*frac*((start+end)/2)/1000
    #print(length)
    #print(len(sliced))
    print('Fraction:', frac)
    print(f'Energy saved [kWh]: {(energy - energy*0.98)*0.15:.3f}', )
    return (energy - energy*0.98) * 0.15
    
result = kde_approx(600, 800)
```

```python
def divide_and_calculate(interval_length):
    result_list = []
    for interval in range(interval_length, 1001, int(interval_length/2)):
        print('Interval:', interval-interval_length, ',', interval)
        result_list.append(kde_approx(interval-interval_length, interval))
        print('-'*25)
    return result_list
result_list = divide_and_calculate(200)
```

```python
plt.plot([x for x in range(len(result_list))], result_list, c='brown')
plt.xticks(ticks=[g for g in range(0, len(result_list)+1, int(len(result_list)/5))], labels=[x for x in range(100, 11001, 100)])
plt.xlabel('Irradiation [W/m2] (interval mid-point)')
plt.ylabel('Energy saved [kWh/m2]')
plt.grid()
```

```python
from math import exp

v = [0, 0.3, 0.4, 0.45, 0.5, 0.51, 0.52, 0.53, 0.54, 0.55, 0.5539, 0.5658, 0.577, 0.59, 0.599, 0.61, 0.62, 0.64, 0.65, 0.66, 0.67, 0.68, 0.69]
c = 5.2/1000 #A/Wm2
k = 1.38065*10**(-23)
e = 1.602*10**(-19)
n = 1
v_t = n*k*(273+25)/e
i_0 = i_1000sc/(exp(v_1000oc/v_t)-1)

def calc_i_d(v_list):
    i_d_list = []
    for v_value in v_list:
        i_d_list.append(i_0*(exp(v_value/(n*v_t))-1))
    return i_d_list

i_d = calc_i_d(v)
```

```python
pv_df = pd.DataFrame({'v': v,
                      'i_d': i_d})
pv_df['i_l_1000'] = i_1000sc - pv_df['i_d']
pv_df['i_l_800'] = i_800sc - pv_df['i_d']

i_600sc = i_800sc - (i_1000sc-i_800sc)

pv_df['i_l_600'] = i_600sc - pv_df['i_d']
pv_df['p_1000'] = pv_df['v'] * pv_df['i_l_1000']
pv_df['p_800'] = pv_df['v'] * pv_df['i_l_800']
pv_df['p_600'] = pv_df['v'] * pv_df['i_l_600']
pv_df

def electrolyzer_p(v_list):
    p_el_list = []
    for v_value in v_list:
        p_el_list.append(v_value*(v_value-0.48)/0.024)
    return p_el_list

pv_df['p_el'] = electrolyzer_p(v)
pv_df
```

```python
pv_df.plot(x='v', y=['p_1000', 'p_800', 'p_600', 'p_el'], marker='.')
plt.ylim([0, 3.2])
plt.xlabel('Voltage [V]')
plt.ylabel('Power [W]')
```
